```python
import numpy as np
```

```python
#Input array
X=np.array([
    [1,0,1,0],
    [1,0,1,1],
    [0,1,0,1]])

#Output
y=np.array([
    [1],
    [1],
    [0]])
```

```python
print(X.shape)
print(y.shape)
```

```python
# Sigmoid function
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
# Derivatives of sigmoid
def dervatives_sigmoid(x):
    return x*(1-x)
```

## Variable initialization

```python
epoch = 5000 # training iterations
lr = 0.1 # Setting Learning rate
inputlayer_neurons = X.shape[1] # number of features in input
hiddenlayer_neurons = 3 # number of hidden layer neurons
output_neurons = 1 # number of neurons at output layer
```

## Weight and bias initialization

```python
wh = np.random.uniform(size=(inputlayer_neurons, hiddenlayer_neurons))
bh = np.random.uniform(size=(1,hiddenlayer_neurons))
wout = np.random.uniform(size=(hiddenlayer_neurons, output_neurons))
bout = np.random.uniform(size=(1, output_neurons))
```

## Iterations

```python
for i in range(epoch):
    # Forward propogation
    hidden_layer_input1 = np.dot(X, wh)
    hidden_layer_input = hidden_layer_input1 + bh
    hidden_layer_activations = sigmoid(hidden_layer_input)
    output_layer_input1 = np.dot(hidden_layer_activations, wout)
    output_layer_input = output_layer_input1 + bout
    output = sigmoid(output_layer_input)
    
    # Backward propogation
    E = y - output
    slope_output_layer = dervatives_sigmoid(output)
    slope_hidden_layer = dervatives_sigmoid(hidden_layer_activations)
    d_output = E * slope_output_layer
    Error_at_hidden_layer = d_output.dot(wout.T)
    d_hidden_layer = Error_at_hidden_layer * slope_hidden_layer
    wout += hidden_layer_activations.T.dot(d_output) * lr
    bout += np.sum(d_output, axis=0, keepdims=True) * lr
    wh += X.T.dot(d_hidden_layer) * lr
    bh += np.sum(d_hidden_layer, axis=0, keepdims=True) * lr
    
    print(output)
```
